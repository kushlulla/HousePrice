### EDA and Cleaning the data 

```python
# lets check the columsn and if we have any missing values
housing_df.info(verbose = True, show_counts=True,)
```

From the data distionary we know that NA stands for not available and not missing value.
so we do not need to alter the data for columns that are object Type

Since we have the value for only 0.5% data for PoolQC we can drop it 

```python
housing_df.drop(["PoolQC","PoolArea"],axis=1,inplace=True)
```

```python
housing_df["MiscFeature"].value_counts()
```

```python
# since we have only a few values in MiscFeature we can change them into a shed yes/no and drop the other data 
housing_df['Shed']=np.where(housing_df['MiscFeature']=='Shed',1,0)
housing_df.drop("MiscFeature",axis=1,inplace=True)
```

```python
housing_df['Fireplaces'].value_counts()
```

```python
#Filling Null values
null_fill=["GarageType","Alley","Fence","FireplaceQu","BsmtFinType1","BsmtFinType2","BsmtQual","BsmtCond","BsmtExposure","GarageFinish","GarageQual","GarageCond","MasVnrType"]
for x in null_fill:
    housing_df[x].fillna("No",inplace=True)
```

```python
#Filling Electrical null values with the most common electrical system
housing_df.Electrical.fillna(housing_df.Electrical.value_counts().index[0],inplace=True)
```

```python
# Lets see if we have any more null values left 
housing_df.columns[housing_df.isna().any()].tolist()
```

```python
# we can fill LotFrontage with median as it is a numeric value
housing_df.LotFrontage.fillna(housing_df.LotFrontage.median(),inplace=True)
```

```python
housing_df[housing_df.MasVnrArea.isna()].MasVnrType
```

```python
# We can see that null values in MasVnrArea are all for Masonry veneer type None so we can fill it with 0
housing_df.MasVnrArea.fillna(0,inplace=True)
```

```python
housing_df[housing_df.GarageYrBlt.isna()].GarageType.value_counts()
```

```python
# all null values are for the houses with no garage so we can fill it with 0
housing_df.GarageYrBlt.fillna(0,inplace=True)
```

### Bivariant analisys with the target variable
To do this lets write a custom function to do the same.


```python
def price_vs(df,col,sc=True):
    if sc:
        sns.scatterplot(data=df,y="SalePrice",x=col)
    else:
        sns.catplot(data=df,x=col,y="SalePrice")
        plt.xticks(rotation=45)
    plt.show()
```

```python
price_vs(housing_df,"Age")
```

Age alone is not a strong predictor of price 

```python
price_vs(housing_df,"OverallCond",False)
```

Overall Condition is a good predictor for the price as the minimum price of a house in good condition is better than the house with poorer condition 

```python
price_vs(housing_df,"LotArea")
```

```python
# Lets drop the outliers of area and see if we get a better pattern 
price_vs(housing_df[housing_df.LotArea<10000],"LotArea")
```

Lot area alone is not a good preditor of price

```python
price_vs(housing_df,"Neighborhood",False)
```

Some Neighborhoods have a higher sale price

```python
price_vs(housing_df,"Condition1",False)
```

Normal is the most common and the price does not show any clear relation 

```python
price_vs(housing_df,"Fireplaces")
```

```python
# Lets see if we can find co-related columns
plt.figure(figsize=(30, 30))
sns.heatmap(housing_df.corr(), annot = True)
```

Above grade (ground) living area has a high corr with sale price lets have a look at it 

```python
price_vs(housing_df,"GrLivArea")
```

```python
corr_val = housing_df[list(housing_df.dtypes[housing_df.dtypes != 'object'].index)].corr()
```

```python
# Lets look at highest co-relation columns  
corr_coef = corr_val[corr_val['SalePrice'] > 0.5]['SalePrice'].sort_values(ascending=False)
print(corr_coef[1:])
corr_coef_cols = [idx for idx in corr_coef.index]
```

```python
plt.figure(figsize=(20, 20))
r=1
for va in corr_coef[1:].index:
    plt.subplot(4,3,r)
    sns.scatterplot(data=housing_df,x=va,y="SalePrice")
    r=r+1
plt.show()
    
```

#### Obseravtions
* Increase in Quality gets better sale price 
* GrLiving Area is a very good predictor of the sale price
* year of remodling isnt a very good preditor for the sale price
